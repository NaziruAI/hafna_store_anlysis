```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('hafna_store.csv')

df.head()
```

```python
sns.countplot(x='Products/services', hue=
              'Date', data = df)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

```python
df['Amount'].plot.hist(bins=10)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cmap='viridis')
```

```python
cdf = df.dropna()
```

```python
cdf.isnull().sum()
```

```python
sns.heatmap(cdf.isnull(), yticklabels=False, cmap='viridis')
```

```python
grouped_data = df.groupby('Products/services')['Amount'].sum().reset_index()

print(grouped_data.sort_values(by='Amount', ascending=False))
```

```python
# Plot the results
plt.figure(figsize=(8, 6))
plt.bar(grouped_data['Products/services'], grouped_data['Amount'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Total Amount')
plt.title('Total Amount per Category')
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.show()
```
